```python
import pandas as pd
```

```python
# 중간 기말 성적.csv

score1 = pd.read_csv(r'C:\Users\User\Desktop\Java_workplace\99_Git\pythonPractice\pythonPractice\py_DataAnalysis\data\중간기말성적.csv',encoding='cp949')
score1.head()
```

* 분반이 A인 데이터 추출

```python
# 성별이 '여'인 데이터 추출
score1[score1['성별']!='여']
```

* 방법 2: 변수.query('조건식')

```python
# 분반이 'B'인 사람들 추출
score1.query('분반=="B"')
```

```python
# '남' 이거나 '중간고사'가 100인 데이터 추출, or
score1.query('성별=="남" | 중간고사 == 100')
```

```python
# 분반 "A" 중에서 '중간고사'와 '기말고사' 80 이상
score1.query('분반=="A" & 중간고사 >=80 & 기말고사 >= 80')
```

* 변수명.query('검색조건')[['column1명','column2명']]
    * [[]] 내에 출력하고자 하는 열 이름 기술

```python
score1.query('분반=="A" & 중간고사 >=80 & 기말고사 >= 80')[['학번','성별']]
```

# 2. 데이터 정렬
* 변수명.sort_values('정렬기준 열이름', ascending=True)
    * ascending 생략 : 오름, =True : 오름, False : 내림차순

```python
# 성별 오름 차순, 기말 기준 내림차순
score1.sort_values(['성별','기말고사'],ascending=[True,False])
```

# 통계함수
* 열이름.통계함수()

```python
# 중간, 기말고사 평균
meandata = score1[['중간고사','기말고사']].mean()
type(meandata)
meandata
```

# Group화
* 변수명.groupby('그룹열이름')[계산할열이름].통계함수()
* Group화 할 열이 2개 이상 => [] 사용

```python
# 분반 기준 '기말고사' 평균 계산
score1.groupby('분반')['기말고사'].mean()
```

```python
# '분반','성별' 기준 중간고사, 기말고사 평균
score1.groupby(['분반','성별'])[['중간고사','기말고사']].mean()
```

# 데이터 결합(열방향)
* 변수명 = pd.merge(변수명1,변수명2, on='결합기준열이름',how='결합방향')
    * 2개의 테이블을 열방향으로 결합
    * 변수명1, 변수명2 : DataFrame 구조의 테이블
    * 결합하고자 하는 테이블들은 반드시 동일한 열이름이 존재해야 함
    * on = '결합기준열이름' : 두개의 테이블 결합시 기준이 되는 열 이름
    -> 일치하지 않으면 결합 X, on = [1,2] : 결합 기준열이 여러개
    * how = 'outer' : 일치하지 않은 결합 기준이면 행으로 추가함

```python
test1 = pd.read_csv(r'C:\Users\User\Desktop\Java_workplace\99_Git\pythonPractice\pythonPractice\py_DataAnalysis\data\test.csv', encoding='EUC-KR')
test1
```

```python
score2 = pd.read_csv(r'C:\Users\User\Desktop\Java_workplace\99_Git\pythonPractice\pythonPractice\py_DataAnalysis\data\출석프로젝트성적.csv', encoding='EUC-KR')
score2.head()
```

```python
# score1, score2 변수 내용을 열방향으로 결합, mgscore1 저장 - 교집합
mgScore1 = pd.merge(score1,score2)
mgScore1
```

```python
#score1, test1 열방향 결합, mgScore2
mgScore2 = pd.merge(score1, test1, on='학번', how='outer')
mgScore2
```

```python
mgScore2 = pd.merge(score1, test1, on='학번', how='right')
mgScore2
```

# 데이터 결합(행방향)
* 변수명 = pd.concat([변수명1, 변수명2])
    * 변수명2를 변수명1에 **행방향**으로 추가

```python
# score1 dataFrame 에서 기말 90 이상이거나 중간 90이상인 기말고사와 중간고사 성별마다 평균 계산
# 결과는 t1 변수에 저장

t1 = score1.query('기말고사 >= 90 | 중간고사 >= 90').groupby('성별')[['중간고사','기말고사']].mean()
t1
```

```python
# score1 dataFrame 에서 기말 90 이상이거나 중간 90이상인 데이터로 성별을 기준 인원수 계산
# 결과는 t2 변수에 저장

t2 = score1.query('기말고사 >= 90 | 중간고사 >= 90').groupby('성별')[['학번']].count()
t2.rename(columns={'중간고사':'인원수'}, inplace=True)
t2
```

```python
t3 = pd.merge(t1,t2, on ='성별')
t3
```
