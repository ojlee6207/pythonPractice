```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='malgun gothic')

data={'이름':['강감찬','홍길동','이순신','이방원'],
      '점수':[50,60,80,65]}

df = pd.DataFrame(data)

# pandas의 plot 이용한 차트 작성
# 데이터프레임.plot()

df.plot(kind='bar', x='이름', rot=0, ylim=[0,100], color='y')

for i in list(df.index):
    
    rate = (df.loc[i]['점수']/df['점수'].sum())*100
    plt.text(x=i-0.1, y=df.loc[i]['점수']-8, s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15, y=df.loc[i]['점수']/2,s=f'{rate:.1f}%')

meanval = df['점수'].mean()
plt.axhline(y=meanval, color = 'r', linestyle='dashed')
plt.text(x=-0.2, y=meanval+1, s=f'평균 : {meanval}점')

plt.xlabel('학생 이름')
plt.ylabel('점수')
plt.title('학생별 시험 점수')

plt.savefig('graph01.png', dpi=400)
print('파일이 저장되었습니다.')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# subplot 작성, matplotlib 이용
data = {'year':[2010,2011,2012,2013,2014],
        'Sales':[500,800,1200,900,1500],
        'Expenses':[300,400,600,500,800]}

df2 = pd.DataFrame(data)

# subplot 생성
fig, axes = plt.subplots(2,2,figsize=(10,9))   # 2행 2열의 subplot(axes), 크기 추가 가능(fig)

# 선그래프 (plot)
axes[0,0].plot(df2['year'],df2['Sales'])
axes[0,0].set_title('Sales', fontsize=14)                    # subplot에서는 set_...로 입력해서 넣어야 함
axes[0,0].set_xlabel('year', fontsize=9)
axes[0,0].set_xticks(df2['year'])

# 막대그래프 (bar)
axes[0,1].bar(df2['year'],df2['Expenses'])
axes[0,1].set_title('Expenses')            

# 산점도 그래프 (scatter)
axes[1,0].scatter(df2['Sales'],df2['Expenses'])
axes[1,0].set_title('Sales vs Expenses')            

# Histogram 그래프
axes[1,1].hist(df2['Sales'], bins=4)            # bins = 
axes[1,1].set_title('Sales Distribution')            


# 그래프 간 간격 조절
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# subplot 작성, seaborn 차트 작성 
data = {'year':[2010,2011,2012,2013,2014],
        'Sales':[500,800,1200,900,1500],
        'Expenses':[300,400,600,500,800]}

df3 = pd.DataFrame(data)

fig, axes = plt.subplots(2,2,figsize=(10,9))

# 선그래프
sns.lineplot(data=df3,x='year',y='Sales',ax=axes[0,0])
axes[0,0].set_title('Sales', fontsize=14)                    # subplot에서는 set_...로 입력해서 넣어야 함
axes[0,0].set_xlabel('year', fontsize=9)
axes[0,0].set_xticks(df2['year'])

# 막대그래프
sns.barplot(data=df3, x='year', y='Expenses', ax=axes[0,1], palette='Set1')

# 산점도 그래프
sns.scatterplot(data=df3, x='Sales', y='Expenses', ax=axes[1,0])

# histogram
sns.histplot(data=df3, x='Sales', ax=axes[1,1])

plt.show()
```
